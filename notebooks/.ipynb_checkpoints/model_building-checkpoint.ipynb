Build a regression model.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import sqlite3
from sqlite3 import Error
from sqlalchemy import create_engine, text
```

```python
# Connect to newly created database and load into a dataframe
with create_engine('sqlite:///../data/hamilton.db/').connect() as conn:
    hamilton_df = pd.read_sql_query(text('SELECT * FROM hamilton_eda'), conn, index_col='index')
```

```python
hamilton_df
```

```python
hamilton_df.describe()
```

```python
# Visualize a pair plot to check histograms for normal distribution and possible relationships between variables
sns.pairplot(data=hamilton_df)
plt.show()
```

```python
# Display a correlation matrix, using Kendall-Tau method since data doesn't seem to follow normal distribution
sns.heatmap(hamilton_df.corr(method='kendall', numeric_only=True), annot=True, cmap='crest')
plt.show()
```

```python
# Define the response and predictor variables, with the bike count at each station will be the response variable
y = hamilton_df['total_bikes']
X = hamilton_df.drop(['citybike_station', 'total_bikes'], axis=1)
```

```python
# Add a constant
X = sm.add_constant(X)
```

```python
# Build a GLM model and fit it
glm_model = sm.GLM(y, X, family=sm.families.Binomial())
glm_results = glm_model.fit()
```

```python
# Build a linear multivariate model and fit it
lin_reg = sm.OLS(y,X)
lin_model = lin_reg.fit()
```

```python
# Build a multinomial regression model and fit it
mn = sm.MNLogit(y,X)
mn_model = mn.fit()
```

```python
# Build a list of linear regression models for each variable and fit it
lr_models = []
for col in X.drop('const', axis=1):
    lr_models.append(sm.OLS(y,X[['const', col]]).fit())
```

```python
# Attempting GAM (Generalized Additive Models)
import pygam as pg
```

```python
y_gam = y.values
X_gam = X.drop('const', axis=1).values
```

```python
gam = pg.LinearGAM(pg.s(0) + pg.s(1) + pg.s(2) + pg.s(3)).fit(X_gam, y_gam)
```

Provide model output and an interpretation of the results. 

#### GLM:

```python
# Print results of the GLM model
print(glm_results.summary())
```

## Based on the log-likelihood of negative infinite, GLM is a poor choice for modeling in this case.

#### Multivariate linear regression:

```python
# Print results of the multivariate linear model
print(lin_model.summary())
```

## The poor R-squared values show poor fitting, so backwards selection will be used to drop features one at a time.

```python
# Backwards feature selection with venue_count removed
X2 = X.drop('venue_count', axis=1)
lin_reg2 = sm.OLS(y,X2)
lin_model2 = lin_reg2.fit()
print(lin_model2.summary())
```

```python
# 2nd backwards feature selection with rating_mean removed
X3 = X2.drop('review_count_mean', axis=1)
lin_reg3 = sm.OLS(y,X3)
lin_model3 = lin_reg3.fit()
print(lin_model3.summary())
```

## Even with two features dropped, the adjusted R-squared value is still poor, meaning that this model is also a poorly fitting one. As the summary says, the high condition number also indicates multiple variables that are codependent on one another, which also makes sense given results in the heatmap of the correlation matrix that was plotted earlier.

#### Multinomial:

```python
# Print the results of the multinomial model
print(mn_model.summary())
```

## The model summaries couldn't even display proper results, meaning that multinomial regression is not a good fit at all. This also makes sense given that multinomial is more appropriate for classification.

#### Linear regression with each variable:

```python
# Print the results of the linear regression model
for lr_model in lr_models:
    print(lr_model.summary())
```

## The low R-squared values for each scenario shows that there is little to no linear relationship between the number of bikes at a station and any of the Yelp data.

#### Generalized additive model for experimentation:

```python
# Print GAM model result summary
gam.summary()
```

## While the pseudo R-squared is higher compared to R-squared of other models, it's still a relatively poor score and not even GAM is a good fit.

# To conclude, there is no straightforward correlation between the number of bike stations at each Citybike station and the average amount of user reviews, the average review score, the average cost, the number of venues nearby and the average distance of venues.
